# ReAct Agent Tutorial

Welcome to this interactive tutorial on building a **ReAct (Reasoning + Acting)** Agent.
We will implement a simple loop where an LLM can "think" about a problem and "act" using tools.

```python
import re

class Agent:
    def __init__(self, llm_mock, tools):
        self.llm = llm_mock
        self.tools = tools
        self.history = []

    def think(self, user_input):
        self.history.append(f"User: {user_input}")
        # In a real scenario, this loops and calls the LLM API
        response = self.llm(user_input)
        return response

print("Agent Class defined.")
```
